```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
final_data = []
counter = 0
for i in os.listdir('data/categorical_data'):
    data = pd.read_csv("data/categorical_data/"+i)
    data['label'] = [counter]*data.shape[0]
    final_data.append(data)
    counter += 1

data = pd.concat(final_data, axis = 0, ignore_index = True)
data
```

```python
for i in data:
    if '_x' not in i:
        if '_y' not in i:
            if 'label' not in i:
                data.drop([i], axis = 1, inplace = True)
```

```python
data
```

```python
labels = data['label']
data.drop(['label'], axis = 1, inplace = True)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, labels,test_size=.2)
```

```python
from lazypredict.Supervised import LazyClassifier

clf = LazyClassifier(verbose=0,ignore_warnings=True, custom_metric=None)
models,predictions = clf.fit(X_train, X_test, y_train, y_test)

print(models)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
clf = ExtraTreesClassifier(n_estimators=100, random_state=0)
clf.fit(X_train, y_train)
```

```python
print("train accuracy = ", int(clf.score(X_train, y_train)*100),"%")
```

```python
print("total cases = ", len(X_test))
right = 0
wrong = 0
y_pred = clf.predict(X_test)
for i in range(len(X_test)):
    if (y_pred[i] == y_test.values[i]):
        right += 1
    else:
        wrong += 1
print("right = ", right)
print("wrong = ", wrong)
print("test accuracy = ", int((len(X_test)/right)*100),"%")
print(clf.score(X_test, y_test))
```

```python
import pickle

# save the model to disk
filename = 'model.sav'
pickle.dump(clf, open(filename, 'wb'))
```

```python
loaded_model = pickle.load(open("model.sav", 'rb'))
result = loaded_model.score(X_test, y_test)
print(result)
```
